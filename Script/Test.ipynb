# Function definitions:

```python
def makedf(fromdf,l,l1):
    to_df= pd.DataFrame()
    for i in l:
        name_of_country = [to_df]
        a = pd.DataFrame(fromdf.loc[i,:])
        b = a.sum(axis = 0, level=0)
        name_of_country.append(b)
        to_df = pd.concat(name_of_country, axis=0, join='outer', ignore_index=False, keys=None,
          levels=None, names=None, verify_integrity=False, copy=True)
    a = pd.DataFrame(fromdf.loc[l1,:])
    l2 = [to_df,a]
    to_df = pd.concat(l2)
    return to_df

def plotdf(df,x="Date",y="No of cases"):
    b = df.T
    plt.figure(); 
    b.plot(figsize=(25, 10)); 
    plt.xlabel(x,fontsize=20);
    plt.ylabel(y,fontsize=20);plt.legend(loc='best')
    plt.show()

def differ(df,lag):
    return df.diff(periods=lag,axis=1)
```

```python
def Consolidate_Datewise(Country):
    def retdf(df):
        datewisedf = df.loc[:,[Country]]
        datewisedf.index.rename(name ="Date",inplace = True)
        datewisedf.columns.name = None
        return datewisedf
    datewise = pd.concat([retdf(Confirmed_HighRisk),retdf(Recovered_HighRisk),retdf(Death_HighRisk)], axis = 1)
    datewise.index =pd.to_datetime(datewise.index)
    datewise.columns = ['Confirmed', 'Recovered', 'Deaths']
    datewise = datewise[datewise.Confirmed != 0]
    datewise["Days Since"]=datewise.index-datewise.index.min()
    return datewise
```

## (1) Polynomial Regression:

```python
def train_poly(train_ml,valid_ml,Category):

    poly = PolynomialFeatures(degree = 8) 
    train_poly=poly.fit_transform(np.array(train_ml["Days Since"]).reshape(-1,1))
    valid_poly=poly.fit_transform(np.array(valid_ml["Days Since"]).reshape(-1,1))
    y=train_ml[Category]
    linreg=LinearRegression(normalize=True)
    linreg.fit(train_poly,y)
    prediction_poly=linreg.predict(valid_poly)
    rmse_poly=np.sqrt(mean_squared_error(valid_ml[Category],prediction_poly))
    model_scores["Polynomial Regression"]= rmse_poly
    print("Root Mean Squared Error for Polynomial Regression: ",rmse_poly) 
    comp_data=poly.fit_transform(np.array(datewise["Days Since"]).reshape(-1,1))
    plt.figure(figsize=(11,6))
    predictions_poly=linreg.predict(comp_data)
    plt.plot(datewise[Category],label="Train Confirmed Cases",linewidth=3)
    plt.plot(datewise.index,predictions_poly, linestyle='--',label="Best Fit for Polynomial Regression",color='black')
    plt.xlabel('Time')
    plt.ylabel(Category+ 'Cases')
    plt.title( Category + "Cases Polynomial Regression Prediction")
    plt.xticks(rotation=90)
    plt.legend()
    def pred_poly():
        new_date=[]
        new_prediction_poly=[]
        for i in range(1,18):
            new_date.append(datewise.index[-1]+timedelta(days=i))
            new_date_poly=poly.fit_transform(np.array(datewise["Days Since"].max()+i).reshape(-1,1))
            new_prediction_poly.append(linreg.predict(new_date_poly)[0])
        model_predictions=pd.DataFrame(zip(new_date,new_prediction_poly),columns=["Date","Polynomial Regression Prediction"])
        return model_predictions
    return pred_poly()
```

## (2) SVM Model:

```python
def train_svm(train_ml,valid_ml,Category):
    svm=SVR(C=1,degree=6,kernel='poly')
    svm.fit(np.array(train_ml["Days Since"]).reshape(-1,1),train_ml["Confirmed"])
    prediction_svm=svm.predict(np.array(valid_ml["Days Since"]).reshape(-1,1))
    rmse_svm=np.sqrt(mean_squared_error(prediction_svm,valid_ml["Confirmed"]))
    model_scores["SVM Model"]= rmse_svm
    print("Root Mean Square Error for SVR Model: ",rmse_svm)
    plt.figure(figsize=(11,6))
    predictions=svm.predict(np.array(datewise["Days Since"]).reshape(-1,1))
    plt.plot(datewise["Confirmed"],label="Train "+Category+" Cases",linewidth=3)
    plt.plot(datewise.index,predictions, linestyle='--',label="Best Fit for SVR",color='black')
    plt.xlabel('Time')
    plt.ylabel('Confirmed Cases')
    plt.title("Confirmed Cases Support Vector Machine Regressor Prediction")
    plt.xticks(rotation=90)
    plt.legend()
    new_date=[]
    new_prediction_svm=[]
    for i in range(1,18):
        new_date.append(datewise.index[-1]+timedelta(days=i))
        new_prediction_svm.append(svm.predict(np.array(datewise["Days Since"].max()+i).reshape(-1,1))[0])

    model_predictions["SVM Prediction"]=new_prediction_svm
    return model_predictions
```

## (3) AR Model:

```python
def train_AR(log_series,train_ml,valid_ml,Category,cliff):
    y_pred=valid_ml.copy()
    model_ar=ARIMA(log_series,(cliff,2,0))
    model_ar_fit=model_ar.fit()
    prediction_ar = model_ar_fit.forecast(len(valid_ml))[0]
    y_pred["AR Model Prediction"]=list(np.exp(prediction_ar))
    rms_ar = np.sqrt(mean_squared_error(list(valid_ml[Category]),np.exp(prediction_ar)))
    model_scores["AR Model"] = rms_ar
    print("Root Mean Square Error for AR Model: ",rms_ar)
    plt.figure(figsize=(10,5))
    plt.plot(train_ml.index,train_ml[Category],label="Train Set",marker='o')
    plt.plot(valid_ml.index,valid_ml[Category],label="Validation Set",marker='*')
    plt.plot(y_pred["AR Model Prediction"],label="AR Model Prediction set",marker="^")
    plt.legend()
    plt.xlabel("Date Time")
    plt.ylabel(Category+" Cases")
    plt.title(Category+" Cases AR Model Forecasting")
    plt.xticks(rotation=90)
    AR_model_new_prediction=[]
    for i in range(1,18):
        AR_model_new_prediction.append(np.exp(model_ar_fit.forecast(len(valid_ml)+i)[0][-1]))
    model_predictions["AR Model Prediction"]=AR_model_new_prediction
    return model_predictions
```

# (4) MA Model:

```python
def train_MA(log_series,train_ml,valid_ml,Category,cliff):
    y_pred=valid_ml.copy()
    model_ma=ARIMA(log_series,(0,2,cliff))
    model_ma_fit=model_ma.fit()
    prediction_ma = model_ma_fit.forecast(len(valid_ml))[0]
    y_pred["MA Model Prediction"]=list(np.exp(prediction_ma))
    rms_ma =np.sqrt(mean_squared_error(list(valid_ml[Category]),np.exp(prediction_ma)))
    model_scores["MA Model"] = rms_ma
    print("Root Mean Square Error for MA Model: ",rms_ma)
    plt.figure(figsize=(10,5))
    plt.plot(train_ml.index,train_ml[Category],label="Train Set",marker='o')
    plt.plot(valid_ml.index,valid_ml[Category],label="Validation Set",marker='*')
    plt.plot(y_pred["MA Model Prediction"],label="MA Model Prediction set",marker="^")
    plt.legend()
    plt.xlabel("Date Time")
    plt.ylabel(Category+" Cases")
    plt.title(Category+" Cases MA Model Forecasting")
    plt.xticks(rotation=90)
    MA_model_new_prediction=[]
    for i in range(1,18):
        MA_model_new_prediction.append(np.exp(model_ma_fit.forecast(len(valid_ml)+i)[0][-1]))
    model_predictions["MA Model Prediction"]=MA_model_new_prediction
    return model_predictions
```
